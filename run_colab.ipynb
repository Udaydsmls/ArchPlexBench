# ArchPlexBench

Trains five sequence model architectures from scratch and compares them on WikiText-2 perplexity.

| Model | Key idea |
|---|---|
| Transformer (decoder-only) | Causal self-attention, Flash Attention |
| Transformer (encoder-decoder) | Bidirectional encoder + causal decoder with cross-attention |
| LSTM | Multi-layer LSTM with weight-tied embeddings |
| Mamba | Selective SSM from scratch, ZOH-discretised diagonal A |
| Mamba (multi-head) | Per-head B/C projections with learnable head mixing |
